```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/data_ndfi_median.csv')

df = df[['ndfi_t0', 'ndfi_t1', 'status']]

dfGrouped = pd.melt(frame=df, value_vars=['ndfi_t0', 'ndfi_t1'], value_name='value', id_vars=['status'])

dfGrouped
```

```python
g = sns.displot(dfGrouped.query('variable == "ndfi_t1"'), 
    x="value", 
    hue="status", 
    kde=True,
    #kind='kde',
    fill=True,
    height=5,
    aspect=2).set(title='Histogram of Validated Samples')

plt.show()
```

```python
g = sns.displot(dfGrouped.query('status == 0'), 
    x="value", 
    hue="variable", 
    kde=True,
    #kind='kde',
    fill=True,
    height=5,
    aspect=2).set(title='Histogram of Rejected Samples')
```
